```python
import pandas as pd

# Assuming df is your DataFrame containing movie data
# Demographic Filtering
C = df["vote_average"].mean()
m = df["vote_count"].quantile(0.9)
print("C: ", C)
print("m: ", m)

new_movies_df = df.copy().loc[df["vote_count"] >= m]

def weighted_rating(x, C=C, m=m):
    v = x["vote_count"]
    R = x["vote_average"]
    return (v/(v + m) * R) + (m/(v + m) * C)

new_movies_df["score"] = new_movies_df.apply(weighted_rating, axis=1)
new_movies_df = new_movies_df.sort_values('score', ascending=False)
print(new_movies_df[["title", "vote_count", "vote_average", "score"]].head(10))

```

```python
# Plotting top 10 movies by popularity
def plot():
    popularity = new_movies_df.sort_values("popularity", ascending=False)
    plt.figure(figsize=(12, 8))
    plt.barh(popularity["title"].head(10), popularity["popularity"].head(10), align="center", color="crimson")
    plt.gca().invert_yaxis()
    plt.title("Top 10 movies by popularity")
    plt.xlabel("Popularity")
    plt.ylabel("Movie Title")
    plt.show()

plot()
```

```python
import json

# Assuming df is your DataFrame containing movie data

# Function to extract genre names
def extract_genre_names(genre_str):
    genres = json.loads(genre_str)
    genre_names = [genre["name"] for genre in genres]
    return genre_names

# Applying the function to the "genres" column
df["genres"] = df["genres"].apply(extract_genre_names)

print(df["genres"][0])  # Check the first entry to verify
```

```python
# Function to extract keyword names
def extract_keyword_names(keyword_str):
    keywords = json.loads(keyword_str)
    keyword_names = [keyword["name"] for keyword in keywords]
    return keyword_names

# Applying the function to the "keywords" column
df["keywords"] = df["keywords"].apply(extract_keyword_names)

print(df["keywords"][0])  # Check the first entry to verify
```

```python
# Function to extract cast names
def extract_cast_names(cast_str):
    cast = json.loads(cast_str)
    cast_names = [actor["name"] for actor in cast]
    return cast_names

# Applying the function to the "cast" column
df["cast"] = df["cast"].apply(extract_cast_names)

print(df["cast"][0])  # Check the first entry to verify
```

```python
# Function to extract crew names
def extract_crew_names(crew_str):
    crew = json.loads(crew_str)
    crew_names = [member["name"] for member in crew]
    return crew_names

# Applying the function to the "crew" column
df["crew"] = df["crew"].apply(extract_crew_names)

print(df["crew"][0])  # Check the first entry to verify
```

```python
df["overview"] = df["overview"].apply(lambda x:x.split())
```

# Remove spaces

```python
df["overview"] = df["overview"].apply(lambda x:[i.replace(" ","") for i in x])
df["genres"] = df["genres"].apply(lambda x:[i.replace(" ","") for i in x])
df["keywords"] = df["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
df["cast"] = df["cast"].apply(lambda x:[i.replace(" ","") for i in x])
df["crew"] = df["crew"].apply(lambda x:[i.replace(" ","") for i in x])

df["release_date"] = df["release_date"].apply(lambda x: [str(x)[:4]])
```

```python
df.head()
```

```python
df["original_title"] = df["original_title"].astype(str)

df["tagline"]=df["tagline"].astype(str)
```

```python
df["text"] = (
    df["release_date"].astype(str)
    + df["overview"].astype(str)
    + df["genres"].astype(str)
    + df["keywords"].astype(str)
    + df["cast"].astype(str)
    + df["crew"].astype(str)
    + df["tagline"].astype(str)
    + df["original_title"].astype(str)
    
    
)
```

```python
df = df[["movie_id", "title", "text"]]
```

```python
df["text"][0]
```

```python
import re

def clean_text(text):
    # Remove unwanted characters
    cleaned_text = text.replace('\'', '').replace('[', '').replace(']', '').replace(',', '')
    # Remove extra spaces
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
    return cleaned_text.strip()  

df["text"]=df["text"].apply(clean_text)
```

```python
df["text"][0]
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
def stem(text):
    ret = []
    words = text.split(" ")
    for word in words:
        ret.append(stemmer.stem(word))
    return " ".join(ret)

df["text"] = df["text"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000, stop_words="english")
data = cv.fit_transform(df["text"]).toarray()
print(data)
```

# Measure Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(data)
sim_df=pd.DataFrame(similarity, index=df['title'], columns=df['title'])
sim_df
```

# Make Recommendations

```python
def make_rocommendation(movie_name):
    recommended_movies = []
    similar_scores = sim_df[movie_name]
    similar_scores = pd.DataFrame(sim_df[movie_name])
    similar_scores = similar_scores.sort_values(movie_name, ascending=False)[1:11]
  
    similar_scores = similar_scores.merge(df, on="title")
    for m in similar_scores["title"]:
        recommended_movies.append(m)
    return recommended_movies

make_rocommendation("The Dark Knight")
```

```python
import pickle
with open('first_model.pkl', 'wb') as f:
    pickle.dump(make_rocommendation, f)
```
